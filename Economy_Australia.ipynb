```python
# STEP 1: Installation
!pip install yfinance --quiet

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import yfinance as yf

print("---Setup complete---")


# WE EXTRACT THE HISTORICAL DATA from YAHOO FINANCE


asx = yf.download("^AXJO", start="2015-01-01", progress=False)
audusd = yf.download("AUDUSD=X", start="2015-01-01", progress=False)

# Safe column extraction to handle inconsistent API column structures.
asx_close = asx["Close"] if "Close" in asx.columns else asx.iloc[:, 0]
audusd_close = audusd["Close"] if "Close" in audusd.columns else audusd.iloc[:, 0]

print("---Data loaded ---")


#  INTEREST RATES (yearly): simplified yearly dataset


interest_rates = pd.DataFrame({
    "Year": list(range(2015, 2026)),
    "Rate": [2.25, 2.00, 1.50, 1.50, 1.00, 0.25, 0.10, 0.85, 3.10, 4.10, 4.35]
})

# ASX GRAPH : VISUAL 1


plt.figure(figsize=(12,5))
plt.plot(asx_close, color="blue")
plt.title("ASX 200 Index (2015–Present)")
plt.grid()
plt.show()

# AUD/USD Conversion : VISUAL 2

plt.figure(figsize=(12,5))
plt.plot(audusd_close, color="green")
plt.title("AUD/USD Exchange Rate")
plt.grid()
plt.show()

# Interest Rate: VISUAL 3

plt.figure(figsize=(10,5))
plt.plot(interest_rates["Year"], interest_rates["Rate"], marker="o")
plt.title("Australia Interest Rates (2015–2025)")
plt.grid()
plt.show()

# ANALYSIS

print("\n|--------------|")
print(" KEY INSIGHTS")
print("|--------------|")

insights = [
    "ASX 200 reflects long-term economic cycles.",
    "Interest rate increases generally slow equity markets.",
    "Lower rates support higher market liquidity.",
    "AUD/USD reflects global risk sentiment and US dollar strength.",
    "Markets respond faster than official economic indicators."
]

for i in insights:
    print("•", i)
print("\nThis was a simple exploration of how Australian macroeconomic indicators interact with financial markets. ")
print("\nAnalysis finished — thanks for following along. ")
```
